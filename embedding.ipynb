```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import pandas as pd
import seaborn as sns
from rich.console import Console
import torch
from torch.nn.functional import cosine_similarity

import archie
```

```python
config = archie.Config(device="cpu")

model, tokenizer = archie.load_model_from_checkpoint(config)
print(model)
```

```python
def get_vector(word):
    toks = tokenizer.encode(word)
    if len(toks) > 1:
        print(f' - {word} is multiple tokens.')
    embedding_vector = model.tok_embeddings.weight[toks].mean(dim=0)
    # print(embedding_vector)
    return embedding_vector


def compare_embedding(w1, w2):
    v1 = get_vector(w1)
    v2 = get_vector(w2)
    # sim = torch.nn.functional.cosine_similarity(
    #     v1.unsqueeze(0),
    #     v2.unsqueeze(0),
    #     dim=1
    # )[0]
    sim = torch.dot(v1, v2)

    print(f'"{w1}" and "{w2}" relate by {sim}')

compare_embedding("Paris", "paris")
compare_embedding("car", " car")
```
